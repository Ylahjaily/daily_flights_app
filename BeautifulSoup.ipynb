```python
def functionX():
    newDict = {}
    
    authorsBrut = soup.select('div.block1 h2')
    citationsBrut = soup.findAll('p', {'class': "texte-h2"})
    
    authorsCleaned = [ segment.string for segment in authorsBrut]
    citationsCleaned = [ segment.string for segment in citationsBrut]
    
    for author in authorsCleaned:
        
        for citation in citationsCleaned:
            #print(citation)
            newDict.update({
                author : citation
            })
    print(newDict)
    
    
functionX()
```

Calculer la moyenne des salaires (en utilisant regex pour récupérer les salaires, évidemment)

```python
salaryBrut = soup.findAll('p', {'class' : 'salaire'})
salaryCleaned = [ segment.string for segment in salaryBrut]

values = []

_somme = 0
for salary in salaryCleaned:
    
    match = re.findall('([0-9]+)', salary)
    
    for m in match:
        intValue = int(m)
        values.append(intValue)
    
    
    for value in values:
        _somme = _somme + value
    
    mean = _somme / len(values)

print(mean)
    
    
```

```python
#Titre du livre
#Notation du livre
#Prix du livre
#Dispo du livre

dataDict = {
    'title' : [],
    'price' : [],
    'notation': [],
    'availability' : []
}

notationBrutList = soup.findAll('p',{'class' : ['star-rating One', 'star-rating Two', 'star-rating Three', 'star-rating Four', 'star-rating Five' ]})


for notation in notationBrutList: 
    if 'One' in str(notation):
        dataDict['notation'].append(1)
    if 'Two' in str(notation):
        dataDict['notation'].append(2)
    if 'Three' in str(notation):
        dataDict['notation'].append(3)
    if 'Four' in str(notation):
        dataDict['notation'].append(4)
    if 'Five' in str(notation):
        dataDict['notation'].append(5)


titlesBrutList = soup.findAll('h3')
titlesCleanedList = [segment.string for segment in titlesBrutList]

priceBrutList = soup.findAll('p', {'class' : 'price_color'})
priceCleanedList = [segment.string for segment in priceBrutList]

for title in titlesCleanedList:
    dataDict['title'].append(title)
    
    
for item in priceCleanedList:
    
    match = re.findall("\d+\.\d+", item)
    for m in match:
        
        dataDict['price'].append(m)
        

availabilityBrutList = soup.findAll('p', {'class': 'instock availability'})


for item in availabilityBrutList:
    
    if '<i class="icon-ok"></i>' in str(item):
        
        dataDict['availability'].append('True')
    else :
        dataDict['availability'].append('False')

dataFrame = pd.DataFrame.from_dict(dataDict)
dataFrame.transpose()

dataFrame.head()
```

```python
mean1stars = dataFrame['notation'] == 1
mean2stars = dataFrame['notation'] == 2
mean3stars = dataFrame['notation'] == 3
mean4stars = dataFrame['notation'] == 4
mean5stars = dataFrame['notation'] == 5

men1starsList = dataFrame[mean1stars]
men2starsList = dataFrame[mean2stars]
men3starsList = dataFrame[mean3stars]
men4starsList = dataFrame[mean4stars]
men5startsList = dataFrame[mean5stars]

mean1 = np.mean([float(i) for i in men1starsList.price])

mean3 = np.mean([float(i) for i in men3startsList.price])

print(mean1)
print(mean3)
```

```python
dictFrame = {
    'name' : [],
    'price' : [],
    'reviews': []
}
```

```python
def getPriceFrame(soup):
    #Chope la séquence voulue
    list_brut = soup.select('div.container.test-site h4.pull-right')
    
    #Découpe en segment la séquence choppée
    list_clean = [segment.string for segment in list_brut]
    
    for item in list_clean:
        matches = re.findall("\d+\.\d+", item)
        
        for match in matches:
            dictFrame['price'].append(float(match))
        


def getTitleFrame(soup):
    #Chope la séquence voulue
    list_brut = soup.select('div.container.test-site a.title')
    
    #Découpe en segment la séquence choppée
    list_clean = [segment.string for segment in list_brut]

    
    for item in list_clean:
        dictFrame['name'].append(str(item))
        

def getReviewFrame(soup):
    #Chope la séquence voulue
    list_brut = soup.select('div.container.test-site p.pull-right')
    
    #Découpe en segment la séquence choppée
    list_clean = [segment.string for segment in list_brut]
    
    for item in list_clean:
        matches = re.findall('([0-9]+)', item)
        
        for match in matches:

            dictFrame['reviews'].append(int(match))
        
getTitleFrame(soup)
getReviewFrame(soup)
getPriceFrame(soup)

print(dictFrame)
```
